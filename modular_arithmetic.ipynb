````python
cipher = "XQFETMHQEAYQRGZIUFTODKBFAMFVTG"
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
def decrypt(text, shift):
    result = ""
    for ch in text:
        if ch.isalpha():
            idx = alphabet.index(ch)
            result += alphabet[(idx - shift) % 26]
        else:
            result += ch
    return result

for s in range(26):
    plain = decrypt(cipher, s)
    print(f"Shift {s}: {plain}")

print(f"Correct decryption is \nShift {12}: {decrypt(cipher, 12)}")
````

````python
def compute_inverse(p):
    inv = [0] * len(p)
    for i, val in enumerate(p):
        inv[val - 1] = i + 1
    return inv

def decrypt(ciphertext, perm):
    m = len(perm)
    inv = compute_inverse(perm)
    if len(ciphertext) % m != 0:
        ciphertext += 'X' * (m - len(ciphertext) % m)
    plaintext = []
    for i in range(0, len(ciphertext), m):
        block = ciphertext[i:i+m]
        plain_block = [''] * m
        for j in range(1, m+1):
            src = inv[j-1] - 1
            plain_block[j-1] = block[src]
        plaintext.extend(plain_block)
    return ''.join(plaintext)

f = [9, 5, 3, 2, 1, 7, 4, 6, 8]
cipher = "ANRTSEAGGYETEMNHOLOINIWGNNMONIEVTSOTBHALPWYOAGNTUOCAIESFEMAHOCESGAWSMRAE"
print(compute_inverse(f))
print(decrypt(cipher, f))
````

__Chosen‑Plaintext Attack on Classical Ciphers__

When an adversary can submit arbitrary plaintexts *x* and obtain the corresponding ciphertexts *y*, the three classical ciphers become trivial to break.

---

### 1. Shift (Caesar) Cipher

*Key*: a single integer `k` in `{0,…,25}` (the shift amount).

*Attack*:\
Choose the plaintext consisting of a single known letter, e.g. `"A"`.\
The ciphertext will be the letter `C = alphabet[(0 + k) % 26]`.\
Thus `k = (index(C) - index('A')) mod 26`.

*Minimum input size*: __1 character__ (any single alphabetic character).

---

### 2. Substitution Cipher

*Key*: a permutation `π` of the 26‑letter alphabet.

*Attack*:\
Submit the entire alphabet as plaintext:

```javascript
plaintext = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

The returned ciphertext reveals the image of each letter under `π`.\
From this we can directly reconstruct the substitution table (the inverse permutation gives decryption).

*Minimum input size*: __26 characters__ (the full alphabet).\
(If the attacker can query individual letters one‑by‑one, a single character per query also works, but the smallest *single* query set that determines the whole key is the full alphabet.)

---

### 3. Vigenère Cipher

*Key*: a word `K` of length `ℓ` (the period).

*Attack*:\
Submit a plaintext consisting of a repeated single letter, e.g. `"AAAA…A"` with length at least `ℓ`.\
The ciphertext will be the key repeated:

```javascript
ciphertext[i] = alphabet[(index('A') + index(K[i mod ℓ])) % 26]
               = K[i mod ℓ]
```

Thus the first `ℓ` characters of the ciphertext give the key directly.

*Minimum input size*: __ℓ characters__, where `ℓ` is the (unknown) key length.\
In the worst case, if the key length is unknown but bounded by `L`, we need `L` characters; otherwise we can first determine `ℓ` (e.g., by trying increasing lengths) and then use `ℓ` characters.
