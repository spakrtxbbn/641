## 2.

The random variable $S$ takes the values $2,\dots ,12$ with the probabilities:

\begin{aligned}
&\Pr[S=2]=\Pr[S=12]=\frac1{36},\\
&\Pr[S=3]=\Pr[S=11]=\frac1{18},\\
&\Pr[S=4]=\Pr[S=10]=\frac1{12},\\
&\Pr[S=5]=\Pr[S=9]=\frac1{9},\\
&\Pr[S=6]=\Pr[S=8]=\frac5{36},\\
&\Pr[S=7]=\frac1{6}.
\end{aligned}

The entropy of $S$ is  

$H(S)= -\sum_{s=2}^{12}\Pr[S=s]\log_2\Pr[S=s].$

Carrying out the sum gives  

\begin{aligned}
H(S) &=
2\!\left(\frac1{36}\log_2\frac{36}{1}\right)
+2\!\left(\frac1{18}\log_2\frac{18}{1}\right)
+2\!\left(\frac1{12}\log_2\frac{12}{1}\right)\\
&\quad+2\!\left(\frac1{9}\log_2\frac{9}{1}\right)
+2\!\left(\frac5{36}\log_2\frac{36}{5}\right)
+\frac1{6}\log_2 6\\[4pt]
&\approx 3.27\ \text{bits}.
\end{aligned}

So the entropy of the sum of two dice is approximately $3.27$ bits.  


## 3.


### 3.1 Plain‑text entropy $H(P)$

$$
\begin{aligned}
H(P) &=-\sum_{x\in\{a,b,c\}} \Pr[x]\log_2\Pr[x] \\[4pt]
     &=-\Bigl(\tfrac12\log_2\tfrac12
               +\tfrac13\log_2\tfrac13
               +\tfrac16\log_2\tfrac16\Bigr) \\[4pt]
     &=0.5+\tfrac13\log_2 3+\tfrac16\log_2 6
       \;\approx\;1.459\ \text{bits}.
\end{aligned}
$$

---

#### 3.2 Key entropy $H(K)$

$$
\begin{aligned}
H(K) &=-\sum_{k\in\{K_1,K_2,K_3\}} \tfrac13\log_2\tfrac13 \\[4pt]
     &=-3\Bigl(\tfrac13\log_2\tfrac13\Bigr)
       =\log_2 3\;\approx\;1.585\ \text{bits}.
\end{aligned}
$$

---

### 3.3 Cipher‑text distribution and entropy $H(C)$

**Encryption matrix**

| Key \ Plain | a | b | c |
|---|---|---|---|
| $K_1$ | 1 | 2 | 3 |
| $K_2$ | 2 | 3 | 4 |
| $K_3$ | 3 | 4 | 1 |

**Prior probabilities**

$$
\Pr[K_i]=\tfrac13,\qquad
\Pr[a]=\tfrac12,\;\Pr[b]=\tfrac13,\;\Pr[c]=\tfrac16 .
$$

**Cipher‑text probabilities**

$$
\begin{aligned}
\Pr[1] &=\tfrac13\!\cdot\!\tfrac12+\tfrac13\!\cdot\!\tfrac16 = \tfrac{2}{9},\\[4pt]
\Pr[2] &=\tfrac13\!\cdot\!\tfrac13+\tfrac13\!\cdot\!\tfrac12 = \tfrac{5}{18},\\[4pt]
\Pr[3] &=\tfrac13\!\cdot\!\tfrac16+\tfrac13\!\cdot\!\tfrac13+\tfrac13\!\cdot\!\tfrac12 = \tfrac13,\\[4pt]
\Pr[4] &=\tfrac13\!\cdot\!\tfrac16+\tfrac13\!\cdot\!\tfrac13 = \tfrac16 .
\end{aligned}
$$

**Entropy**

$$
\begin{aligned}
H(C) &=-\sum_{y\in\{1,2,3,4\}} \Pr[y]\log_2\Pr[y] \\[4pt]
     &=-\Bigl(\tfrac{2}{9}\log_2\tfrac{2}{9}
               +\tfrac{5}{18}\log_2\tfrac{5}{18}
               +\tfrac13\log_2\tfrac13
               +\tfrac16\log_2\tfrac16\Bigr) \\[4pt]
     &\approx\;1.954\ \text{bits}.
\end{aligned}
$$

---

### 3.4 Conditional key entropy $H(K\mid C)$

#### 3.4.1 Joint probabilities $\Pr[K \text{ and } C=y]$

$$
\begin{aligned}
\Pr[K_1\!\&\!C=1] &=\tfrac13\!\cdot\!\tfrac12 = \tfrac16, &
\Pr[K_3\!\&\!C=1] &=\tfrac13\!\cdot\!\tfrac16 = \tfrac1{18},\\[4pt]
\Pr[K_1\!\&\!C=2] &=\tfrac13\!\cdot\!\tfrac13 = \tfrac19, &
\Pr[K_2\!\&\!C=2] &=\tfrac13\!\cdot\!\tfrac12 = \tfrac16,\\[4pt]
\Pr[K_1\!\&\!C=3] &=\tfrac13\!\cdot\!\tfrac16 = \tfrac1{18}, &
\Pr[K_2\!\&\!C=3] &=\tfrac13\!\cdot\!\tfrac13 = \tfrac19,\\[4pt]
\Pr[K_3\!\&\!C=3] &=\tfrac13\!\cdot\!\tfrac12 = \tfrac16, &
\Pr[K_2\!\&\!C=4] &=\tfrac13\!\cdot\!\tfrac16 = \tfrac1{18},\\[4pt]
\Pr[K_3\!\&\!C=4] &=\tfrac13\!\cdot\!\tfrac13 = \tfrac19 .
\end{aligned}
$$

#### 3.4.2 Conditional probabilities $\Pr[K=k\mid C=y]$

Using $\Pr[K=k\mid C=y]=\dfrac{\Pr[K=k\!\&\!C=y]}{\Pr[C=y]}$:

$$
\begin{aligned}
\Pr[K_1\mid C=1] &=\frac{\tfrac16}{\tfrac29}= \tfrac34, &
\Pr[K_3\mid C=1] &=\frac{\tfrac1{18}}{\tfrac29}= \tfrac14,\\[4pt]
\Pr[K_1\mid C=2] &=\frac{\tfrac19}{\tfrac5{18}}= \tfrac25, &
\Pr[K_2\mid C=2] &=\frac{\tfrac16}{\tfrac5{18}}= \tfrac35,\\[4pt]
\Pr[K_1\mid C=3] &=\frac{\tfrac1{18}}{\tfrac13}= \tfrac16, &
\Pr[K_2\mid C=3] &=\frac{\tfrac19}{\tfrac13}= \tfrac13,\\[4pt]
\Pr[K_3\mid C=3] &=\frac{\tfrac16}{\tfrac13}= \tfrac12, &
\Pr[K_2\mid C=4] &=\frac{\tfrac1{18}}{\tfrac16}= \tfrac13,\\[4pt]
\Pr[K_3\mid C=4] &=\frac{\tfrac19}{\tfrac16}= \tfrac23 .
\end{aligned}
$$

#### 3.4.3 Entropy for each ciphertext

$$
\begin{aligned}
H(K\mid1) &= -\tfrac34\log_2\tfrac34-\tfrac14\log_2\tfrac14 \;\approx\;0.811,\\[4pt]
H(K\mid2) &= -\tfrac25\log_2\tfrac25-\tfrac35\log_2\tfrac35 \;\approx\;0.971,\\[4pt]
H(K\mid3) &= -\tfrac16\log_2\tfrac16-\tfrac13\log_2\tfrac13-\tfrac12\log_2\tfrac12 \;\approx\;1.459,\\[4pt]
H(K\mid4) &= -\tfrac13\log_2\tfrac13-\tfrac23\log_2\tfrac23 \;\approx\;0.918 .
\end{aligned}
$$

#### 3.4.4 Overall conditional key entropy

$$
\begin{aligned}
H(K\mid C) &= \sum_{y}\Pr[C=y]\;H(K\mid y) \\[4pt]
           &= \tfrac{2}{9}(0.811)
            + \tfrac{5}{18}(0.971)
            + \tfrac13(1.459)
            + \tfrac16(0.918) \\[4pt]
           &\approx\;1.089\ \text{bits}.
\end{aligned}
$$

### 3.5 Conditional plaintext entropy $H(P\mid C)$

The calculation mirrors the key case because each ciphertext uniquely determines a pair $(\text{key},\text{plaintext})$.  
We therefore reuse the same conditional probabilities, now interpreted as $\Pr[P=p\mid C=y]$.

#### 3.5.1 Conditional probabilities $\Pr[P\mid C]$, can reuse from 3.4

| $y$ | $\Pr[a\mid y]$ | $\Pr[b\mid y]$ | $\Pr[c\mid y]$ |
|---|---|---|---|
| 1 | $\frac34$ | $0$ | $\frac14$ |
| 2 | $\frac35$ | $\frac25$ | $0$ |
| 3 | $\frac12$ | $\frac13$ | $\frac3{18}$ |
| 4 | $0$ | $\frac23$ | $\frac13$ |

#### 3.5.2 Entropy for each ciphertext

$$
\begin{aligned}
H(P\mid1) &= -\frac34\log_2 \frac34 -\frac14\log_2 \frac14 \;\approx\;0.811,\\[4pt]
H(P\mid2) &= -\frac35\log_2 \frac35 -\frac25\log_2 \frac25 \;\approx\;0.971,\\[4pt]
H(P\mid3) &= -\frac12\log_2 \frac12 -\frac13\log_2 \frac13 -\frac3{18}\log_2 \frac3{18} \;\approx\;1.459,\\[4pt]
H(P\mid4) &= -\frac23\log_2 \frac23 -\frac13\log_2 \frac13 \;\approx\;0.918 .
\end{aligned}
$$

#### 3.5.3 Overall conditional plaintext entropy

$$
\begin{aligned}
H(P\mid C) &= \sum_{y}\Pr[C=y]\;H(P\mid y) \\[4pt]
           &= \tfrac{2}{9}(0.811)
            + \tfrac{5}{18}(0.971)
            + \tfrac13(1.459)
            + \tfrac16(0.918) \\[4pt]
           &\approx\;1.089\ \text{bits}.
\end{aligned}
$$
