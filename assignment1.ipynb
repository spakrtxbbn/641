```python
cipher = "XQFETMHQEAYQRGZIUFTODKBFAMFVTG"
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
def decrypt(text, shift):
    result = ""
    for ch in text:
        if ch.isalpha():
            idx = alphabet.index(ch)
            result += alphabet[(idx - shift) % 26]
        else:
            result += ch
    return result

for s in range(26):
    plain = decrypt(cipher, s)
    print(f"Shift {s:2d}: {plain}")

print(f"Correct decryption is \nShift {12}: {decrypt(cipher, 12)}")
```

```python
def compute_inverse(p):
    inv = [0] * len(p)
    for i, val in enumerate(p):
        inv[val - 1] = i + 1
    return inv

def decrypt(ciphertext, perm):
    m = len(perm)
    inv = compute_inverse(perm)
    if len(ciphertext) % m != 0:
        ciphertext += 'X' * (m - len(ciphertext) % m)
    plaintext = []
    for i in range(0, len(ciphertext), m):
        block = ciphertext[i:i+m]
        plain_block = [''] * m
        for j in range(1, m+1):
            src = inv[j-1] - 1
            plain_block[j-1] = block[src]
        plaintext.extend(plain_block)
    return ''.join(plaintext)

f = [9, 5, 3, 2, 1, 7, 4, 6, 8]
cipher = "ANRTSEAGGYETEMNHOLOINIWGNNMONIEVTSOTBHALPWYOAGNTUOCAIESFEMAHOCESGAWSMRAE"
print("a. ", compute_inverse(f))
print("b. ", decrypt(cipher, f))
```

## 4

### 4.1 Shift Cipher

Choose the plaintext consisting of a single known letter, for example `"A"`.\
The ciphertext will be the letter `C = alphabet[(0 + k) % 26]`.\
Thus `k = (index(C) - index('A')) mod 26`.

Only 1 character from the cipher's alphabet is needed as a minimum input.

---

### 4.2 Substitution Cipher

Submit the entire alphabet as plaintext, for example `"ABCDEFGHIJKLMNOPQRSTUVWXYZ"`
The returned ciphertext reveals the mapping of each letter .\
From this we can directly reconstruct the substitution table (the inverse permutation gives decryption).

Only 26 characters, or the full alphabet is needed as a minimum input.\

---

### 4.3 Vigenere Cipher

Submit a plaintext consisting of a repeated single letter, for example `"AAAA…A"` with a length at least some length `l`.\
The ciphertext will be the key repeated:

The first `l` characters of the ciphertext give the key directly.

Where `l` is the unknown key length, you only need at least `l` characters as a minimum input.\

```python
from collections import Counter

ENGLISH_FREQ = [
    8.167,  # A
    1.492,  # B
    2.782,  # C
    4.253,  # D
    12.702, # E
    2.228,  # F
    2.015,  # G
    6.094,  # H
    6.966,  # I
    0.153,  # J
    0.772,  # K
    4.025,  # L
    2.406,  # M
    6.749,  # N
    7.507,  # O
    1.929,  # P
    0.095,  # Q
    5.987,  # R
    6.327,  # S
    9.056,  # T
    2.758,  # U
    0.978,  # V
    2.360,  # W
    0.150,  # X
    1.974,  # Y
    0.074   # Z
]

def ic_of(text): # find the index of coincidence
    n = len(text)
    if n <= 1:
        return 0.0
    freq = Counter(text)
    numerator = sum(v * (v - 1) for v in freq.values())
    denominator = n * (n - 1)
    return numerator / denominator


def avg_ic_for_keylen(text, k):
    ics = []
    for i in range(k):
        column = text[i::k] # every k‑th character, starting at i
        ics.append(ic_of(column))
    return sum(ics) / k, ics # return average of ics and list of k ics

def mg_score(counts, shift):
    total = 0.0
    for i in range(26): # for every letter
        idx = (i + shift) % 26 # shift every letter
        total += ENGLISH_FREQ[i] * counts[idx] # multiply original letter freq with shifted letter count
    return total

def best_shift_for_column(col):
    n = len(col)
    # count the letters
    counts = [0] * 26
    for c in col:
        counts[ord(c) - ord('A')] += 1

    best_shift = None
    best_score = -1.0
    for shift in range(26): # test every letter shift
        score = mg_score(counts, shift)
        if score > best_score:
            best_score = score
            best_shift = shift

    # Convert the raw numerator into the familiar M_g = score / n
    mg_value = best_score / n
    return best_shift, mg_value


def vigenere_decrypt(cipher, key):
    plain = []
    klen = len(key)
    for i, c in enumerate(cipher):
        k = ord(key[i % klen]) - ord('A')
        p = (ord(c) - ord('A') - k) % 26
        plain.append(chr(p + ord('A')))
    return ''.join(plain)



cipher = "BNWEOMZOBNTALBAECJHKXSFGYZXMBNBVZSBXMETRRTNOLTUWGXKOTJTUFIAYFCLTOWMPXPEGVMLSBXICMTVFZBAEZAGMWSTWKLBJXKMZTDRTNOZIAYBAMWVUXILHNJWILWEAMQGGGZXKHDRAGBAESAKAMPYSVMMHRJXNHRRAYGHUJJBBXTUWVWWENKVTXVRJEGTSCGLABBYWRWNAEWUGWESAGQMIBFGWMSZSKBXNBMZPMOQWUCZIGTKQTNXWKVBGUSGBHEEJBAAUZSGJNTGGKMTLYQYWNLGZBVZSHHRWNNRWWIVOZHNBXRCSNTXHEDBKAISQHCYIAVMPTTLGNZXSCWGLBNTSEUHSGSETROHJMQFEBFMPXOEQLBTRGLNZGIAYLWFENLMMGTVGGBHPESVBBCNDMPBNTKBBPIYDBUIRBNXGHUELAMHRVWLQYYBMYQGDGZTBROHJXAIEAVBVZAYEHAMAYDRWNRGAFMHNCJTKMIPWLBTRGLNZGIAYLWFENLMMGTVGGBHTUWHZXTVUTTMHVFZABTJAETBMCJHDXYBMKXKAPLBKXDBFTTWKAMMP"

global_ic = ic_of(cipher)
print(f"Overall Index of Coincidence (IC) = {global_ic:.5f}\n")

max_keylen = 10 # 10 trials
ic_table = [] # (k, avg_ic, [ic_of_each_column])
print("k   avg IC   column ICs")
print("-" * 40)
for k in range(1, max_keylen + 1): # populate the ic table
    avg_ic, col_ics = avg_ic_for_keylen(cipher, k)
    ic_table.append((k, avg_ic, col_ics))
    col_ics_str = " ".join(f"{c:.3f}" for c in col_ics)
    print(f"{k:2d}  {avg_ic:.3f}   {col_ics_str}")

target_ic = 0.065
best = min(ic_table, key=lambda t: abs(t[1] - target_ic)) # find the closest matching ic
key_len = best[0]
print("\nEstimated key length = ", key_len)

key_shifts = [] # store the key letters
print("\nM_g values:")
print("-" * 40)
for i in range(key_len): # for every column
    column = cipher[i::key_len]
    shift, mg_value = best_shift_for_column(column) # find the best shift
    key_shifts.append(shift)
    print(f"column {i} : shift = {shift:2d}  (M_g = {mg_value:.2f})")

key = ''.join(chr(s + ord('A')) for s in key_shifts) # add the offset to the key
print("\nKey:\n", key)

plaintext = vigenere_decrypt(cipher, key)
print("\nPlaintext:\n", plaintext)
```
