```python
import math
import sys
from typing import Tuple

def pollard_rho(N, start = 31):
    """
    Returns (factor, iterations) where *factor* is a non‑trivial divisor of N.
    The iteration counter counts how many times the loop body
    (the “hare” step) is executed.
    """
    if N % 2 == 0:
        return 2, 1                     # trivial even case

    # deterministic polynomial f(x) = x² + 1 (mod N)
    f = lambda x: (pow(x, 2, N) + 1) % N

    # initialise the tortoise and the hare
    tortoise = start % N
    hare     = f(tortoise)            # hare = f(x1)
    iterations = 1                    # we have already evaluated f once

    while True:
        # move the two pointers
        tortoise = f(tortoise)                # 1 step
        hare     = f(f(hare))                 # 2 steps
        iterations += 2                        # two evaluations of f

        d = math.gcd(abs(hare - tortoise), N)
        if d == 1:
            continue                # still no factor – keep looping
        if d == N:
            # failure – the algorithm fell into a cycle that only yields N.
            # In a deterministic setting we simply restart with a different
            # start value (here we add 1 to the start and try again).
            return pollard_rho(N, start + 1)
        # a proper factor has been found
        return d, iterations


# ------------------------------------------------------------
# 2.  Helper that returns the *complete* factorisation
# ------------------------------------------------------------
def factor(N):
    """
    Returns (p, q, iters) where N = p·q (p ≤ q) and *iters* is the number of
    iterations that pollard_rho needed to discover the first factor.
    """
    p, iters = pollard_rho(N, start=31)
    q = N // p
    if p > q:                     # keep the smaller factor first – nicer output
        p, q = q, p
    return p, q, iters



numbers = [
    53081719,
    44818676050679,
    14690966543846720848264259950499,
]

for N in numbers:
    p, q, iters = factor(N)
    print(f"N = {N}")
    print(f"  factor 1 = {p}")
    print(f"  factor 2 = {q}")
    print(f"  iterations needed (Pollard_rho, x_1=31) = {iters}\n")
```

```python
import math
import itertools

def get_exponent_vector(val, base):
    counts = {p: 0 for p in base}
    
    # Handle the -1 in the factor base if negative residues were used
    if val < 0:
        counts[-1] = 1
        val = -val
        
    for p in base:
        if p == -1:
            continue
        while val % p == 0:
            counts[p] += 1
            val //= p
            
    if val == 1:
        return counts
    return None

n = 347881
base = [-1, 2, 3, 5, 7, 11, 13, 17, 19, 23]

print(f"Target n = {n}")
print(f"Factor base = {base}\n")

# Find B-smooth values of z^2 mod n
smooth_relations = []

# Testing strictly within the range 600 <= z <= 660
for z in range(600, 661):
    r = (z**2) % n
    
    # Test if the residue is smooth over our factor base
    vector = get_exponent_vector(r, base)
    if vector is not None:
        smooth_relations.append((z, r, vector))
        print(f"Found smooth relation for z={z}: {z}^2 mod n = {r}")
        
print(f"\nTotal smooth relations found: {len(smooth_relations)}\n")

# Find a subset of relations whose exponents sum to even numbers
# test combinations of increasing sizes
for r_len in range(1, len(smooth_relations) + 1):
    for combo in itertools.combinations(smooth_relations, r_len):
        vector_sum = {p: 0 for p in base}
        
        # Sum the exponent vectors for this combination
        for z, r, counts in combo:
            for p in base:
                vector_sum[p] += counts[p]
        
        # Check if we have a linear dependency modulo 2 (all sums are even)
        if all(v % 2 == 0 for v in vector_sum.values()):
            
            # Compute x and y for the congruence x^2 ≡ y^2 (mod n)
            x = 1
            for z, r, counts in combo:
                x = (x * z) % n
                
            y = 1
            for p in base:
                if p == -1:
                    continue
                # y is the product of primes raised to half of their summed exponents
                y = (y * pow(p, vector_sum[p] // 2, n)) % n
                
            # Calculate the greatest common divisors to extract factors
            factor1 = math.gcd((x - y) % n, n)
            factor2 = math.gcd((x + y) % n, n)
            
            # Verify that the factors are non-trivial (not 1 and not n)
            if 1 < factor1 < n:
                print("CONGRUENCE FOUND")
                print(f"Used subset of size {r_len}: {[c[0] for c in combo]}")
                print(f"x = {x}, y = {y}")
                print(f"x^2 mod n = {(x**2) % n}")
                print(f"y^2 mod n = {(y**2) % n}")
                print("\nFACTORIZATION")
                print(f"Factors of {n} are {factor1} and {factor2}")
                print(f"Verification: {factor1} * {factor2} = {factor1 * factor2}")
                break
```

```python
def chinese_remainder_theorem(remainders, moduli):
    total_sum = 0
    prod = 1
    for n in moduli:
        prod *= n
        
    for n_i, a_i in zip(moduli, remainders):
        p = prod // n_i
        total_sum += a_i * modinv(p, n_i) * p
        
    return total_sum % prod

def integer_cbrt(n):
    low = 0
    high = n
    while low < high:
        mid = (low + high) // 2
        if mid**3 < n:
            low = mid + 1
        else:
            high = mid
    return low


moduli = [25777, 22879, 66277]

# Ciphertexts for Message 1
c1_rems = [19052, 4546, 44619]

# Ciphertexts for Message 2
c2_rems = [1708, 11733, 19731]

# Apply Chinese Remainder Theorem to find m^3
M1_cubed = chinese_remainder_theorem(c1_rems, moduli)
M2_cubed = chinese_remainder_theorem(c2_rems, moduli)

# Take the integer cube root to find the actual message integers
m1_int = integer_cbrt(M1_cubed)
m2_int = integer_cbrt(M2_cubed)

print(f"Decoded Integer for Block 1: {m1_int}")
print(f"Decoded Integer for Block 2: {m2_int}")

# Convert the integers back into 3-character blocks
block1 = int_to_block(m1_int)
block2 = int_to_block(m2_int)

print(f"Message 1 block: {block1}")
print(f"Message 2 block: {block2}")
print(f"Full 6-character message: {block1}{block2}")
```
